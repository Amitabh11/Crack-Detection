```python
import skimage.color
import skimage.filters
import skimage.io

image = skimage.io.imread(fname='/content/drive/MyDrive/key1_o.jpg')
skimage.io.imshow(image)

blur = skimage.color.rgb2gray(image)
blur = skimage.filters.gaussian(blur, sigma=2)

t = skimage.filters.threshold_otsu(blur)
mask = blur > t

sel = np.zeros_like(image)
sel[mask] = image[mask]

# display the result
skimage.io.imshow(sel)
```

```python
for i in range(229):
  for j in range(410):
    for k in range(3):
      if sel[i][j][k] != 0:
        sel[i][j][k] = 255
```

```python
plt.imshow(sel)
```

```python
image1 = skimage.io.imread(fname='/content/drive/MyDrive/key1_r.jpg')
skimage.io.imshow(image1)

blur1 = skimage.color.rgb2gray(image1)
blur1 = skimage.filters.gaussian(blur1, sigma=2)

t1 = skimage.filters.threshold_otsu(blur1)
mask1 = blur1 > t1

sel1 = np.zeros_like(image1)
sel1[mask1] = image1[mask1]

# display the result
skimage.io.imshow(sel1)
```

```python
sel1.shape
```

```python
for i in range(229):
  for j in range(410):
    for k in range(3):
      if sel1[i][j][k] != 0:
        sel1[i][j][k] = 255
```

```python
plt.imshow(sel1)
```
