```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img_cny=cv2.Canny(img_gray,35,125)
```

```python
plt.imshow(img_cny,cmap='gray')
```

```python
# find the contours from the thresholded image
contours= cv2.findContours(img_cny, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)[0]
# draw all contours
image = cv2.drawContours(img, contours, -1, (0, 255, 0), 2)
```
