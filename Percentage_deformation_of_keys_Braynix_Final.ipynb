```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
from skimage.morphology import (erosion, dilation, opening, closing,  # noqa
                                )
import matplotlib.pyplot as plt

```

```python
def plot_comparison(original, filtered, filter_name):

    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(32,16), sharex=True,
                                   sharey=True)
    ax1.imshow(original, cmap="gray")
    ax1.set_title('original')
    ax1.axis('off')
    ax2.imshow(filtered, cmap="gray")
    ax2.set_title(filter_name)
    ax2.axis('off')
```

```python
image = cv2.imread('/content/key1.jpg')
shifted = cv2.pyrMeanShiftFiltering(image, 30, 18)

gray = cv2.cvtColor(shifted, cv2.COLOR_BGR2GRAY)
#blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Display original image
cv2_imshow(image)
cv2_imshow(gray)

```

```python
gray = cv2.cvtColor(shifted, cv2.COLOR_BGR2GRAY)

ret, thresh1 = cv2.threshold(shifted, 5, 255, cv2.THRESH_BINARY_INV)
gr = cv2.cvtColor(thresh1, cv2.COLOR_BGR2GRAY)

cv2_imshow(thresh1)
#cv2_imshow(gr)
```

```python
open = closing(opening(gr))
#gr1 = cv2.cvtColor(open, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(open, 100, 255, cv2.THRESH_BINARY)

cv2_imshow(open)
cv2_imshow(thresh)
```

```python
c = 0
cnt = 0
l = thresh.shape[1]//2

for i in thresh:
    i = i.tolist()
    fh = i[:l]
    sh = i[l:]
    if 255 in i:
        c+= 1
    if 255 in fh and 255 in sh:
        cnt += 1        
        x1 = fh.index(255)
        y1 = len(fh) - 1 - fh[::-1].index(255)
        wid1 = y1-x1
        x2 = sh.index(255)
        y2 = len(sh) - 1 - sh[::-1].index(255)
        wid2 = y2-x2
        #print("wid-1 = {}, wid-2 = {}".format(wid1,wid2))
        print("difference = ",abs(wid2-wid1))
print("len/2 = ",thresh.shape)
print("total identified rows = {}, keys in both rows = {}".format(c,cnt))
```
